```python
class Animal:
    def __init__(self, species):
        self.species = species

    def show_info(self):
        print(f"tür: {self.species}")

        
class Mammal(Animal):
    def __init__(self, species, weight, speed):
        super().__init__(species)
        self.weight = weight
        self.speed = speed

    def show_info(self):
        super().show_info()
        print(f"Weight: {self.weight} kg")
        print(f"Speed: {self.speed} km/h")

        
class Bird(Animal):
    def __init__(self, species, wingspan):
        super().__init__(species)
        self.wingspan = wingspan

    def show_info(self):
        super().show_info()
        print(f"Wingspan: {self.wingspan} cm")

        
class Reptile(Animal):
    def __init__(self, species, length):
        super().__init__(species)
        self.length = length

    def show_info(self):
        super().show_info()
        print(f"Length: {self.length} cm")


class Zookeeper:
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id
        self.assigned_animal = None

    def assign_animal(self, animal):
        self.assigned_animal = animal
        print(f"{self.name} bu hayvana atandı:")
        animal.show_info()


class Zoo:
    def __init__(self):
        self.zookeepers = []

    def add_zookeeper(self, zookeeper):
        self.zookeepers.append(zookeeper)

    def show_zookeepers(self):
        for zookeeper in self.zookeepers:
            print(f"Zookeeper adı: {zookeeper.name}, ID: {zookeeper.employee_id}")
            if zookeeper.assigned_animal:
                zookeeper.assigned_animal.show_info()
            else:
                print("No animal assigned")



lion = Mammal(species="Lion", weight=190, speed=50)
eagle = Bird(species="Eagle", wingspan=200)
snake = Reptile(species="Snake", length=150)


arda = Zookeeper(name="arda kamçı", employee_id=1)
emir = Zookeeper(name="emir can", employee_id=2)


arda.assign_animal(lion)
emir.assign_animal(eagle)


zoo = Zoo()
zoo.add_zookeeper(arda)
zoo.add_zookeeper(emir)


zoo.show_zookeepers()
```
