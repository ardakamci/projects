```python
def urun_ekle(alısveris_listesi):
    urun_adı=input("ürün adı:")
    miktar = int(input("miktarı: "))
    fiyat =int(input("birim fiyatı: "))
    urun={"isim":urun_adı,"miktar":miktar,"birim fiyatı":fiyat}
    alısveris_listesi.append(urun)
    print(f"{urun_adı} eklendi.")
    return alısveris_listesi

def urun_guncelle(alısveris_listesi):
    urun_adı = input("adı güncellenecek ürünün adı: ")
    yeni_miktar = int(input("yeni birim miktar: "))
    yeni_fiyat = int(input("Yeni fiyat: "))

    for urun in alısveris_listesi:
        if urun['isim'] == urun_adı:
            urun['miktar'] = yeni_miktar
            urun['fiyat'] = yeni_fiyat
            print(f"{urun_adı} için ürün güncellendi.")
            return alısveris_listesi
    print(f"{urun_adı} bulunamadı.")
    return alısveris_listesi

def urun_sil(alısveris_listesi):

    urun_adı = input("Silinecek ürünün adı: ")
    alısveris_listesi = [urun for urun in alısveris_listesi if urun['isim'] != urun_adı]
    print(f"{urun_adı} listeden çıkarıldı.")
    return alısveris_listesi

def urunleri_goruntule(alısveris_listesi):

    if not alısveris_listesi:
        print("Liste boş.")
        return
    print("ürün Listesi:")
    for urun in alısveris_listesi:
        print(f"ürün: {urun['isim']}, miktar: {urun['miktar']},fiyat:{urun['birim fiyatı']}")

def toplam_maliyet(alısveris_listesi):
    toplam = sum(urun['miktar'] * urun['birim fiyatı'] for urun in alısveris_listesi)
    print(f"toplam Maliyet: {toplam:.2f}")
    return toplam

alısveris_listesi = []
while True:
    print("1. ürün Ekle")
    print("2. ürün Güncelle")
    print("3. ürün Sil")
    print("4. ürünleri Görüntüle")
    print("5. toplam maliyet")
    print("6. Çıkış")

    secim = input("Bir seçenek girin (1-6): ")

    if secim == "1":
        alısveris_listesi =urun_ekle(alısveris_listesi)
    elif secim == "2":
         alısveris_listesi = urun_guncelle(alısveris_listesi)
    elif secim == "3":
         alısveris_listesi = urun_sil(alısveris_listesi)
    elif secim == "4":
       urunleri_goruntule(alısveris_listesi)
    elif secim == "5":
        toplam_maliyet(alısveris_listesi)
    elif secim == "6":
        print("Çıkış yapılıyor...")
        break
    else:
        print("Geçersiz seçenek, lütfen tekrar deneyin.")
```
